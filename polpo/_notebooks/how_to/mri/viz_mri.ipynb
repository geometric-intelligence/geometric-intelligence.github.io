# How to visualize MRI data?

```python
from polpo.plot.mri import MriPlotter, MriSlicer, SlicesPlotter
from polpo.preprocessing import ListSqueeze
from polpo.preprocessing.load.pregnancy import PregnancyPilotMriLoader
from polpo.preprocessing.mri import MriImageLoader
```

## Loading data

We'll create a pipeline that:
* [downloads data](load_figshare_data.ipynb) from [figshare](https://figshare.com/articles/dataset/pregnancy-data/28339535)
* finds [nii](https://en.wikipedia.org/wiki/Neuroimaging_Informatics_Technology_Initiative) file in folder
* loads image as an array

```python
SESSION_ID = 1
```

```python
loader = PregnancyPilotMriLoader(subset=[SESSION_ID])

pipe = loader + ListSqueeze() + MriImageLoader()
```

```python
img_fdata = pipe()
```

## Visualization

`MriSlicer` slices the MRI image.
`SlicesPlotter` takes care of plotting slices. By default, it plots side, front and top views.


```python
slices_plotter = SlicesPlotter()
slicer = MriSlicer()
```

```python
slices_plotter.plot(slicer.slice(img_fdata, (35, 70, 105)));
```

For convenience, there's also `MriPlotter`.

```python
mri_plotter = MriPlotter()
```

```python
mri_plotter.plot(img_fdata, [val // 2 for val in img_fdata.shape]);
```
